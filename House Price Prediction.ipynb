```python
import pandas as pd
import streamlit as st
import joblib

model = joblib.load(
    r"C:\Users\rites\AppData\Local\Packages\5319275A.WhatsAppDesktop_cv1g1gvanyjgm\LocalState\sessions\E5E974812420014CBE06FD5F226D3B37E07DBD24\transfers\2026-03\xgb_model.jb"
)

st.title("üè† House Price Prediction")
st.write("Enter house details below")

# üî• Use model features directly
inputs = model.feature_names_in_.tolist()

input_data = {}
for feature in inputs:
    if feature == 'CentralAir':
        input_data[feature] = st.selectbox(feature, ['Yes', 'No'])
    else:
        input_data[feature] = st.number_input(feature, value=0.0)

if st.button("Predict Price"):
    input_data['CentralAir'] = 1 if input_data['CentralAir'] == 'Yes' else 0

    input_df = pd.DataFrame([input_data])
    input_df = input_df.astype(float)

    prediction = model.predict(input_df.values)

    st.success(f"üí∞ Predicted House Price: ${prediction[0]:,.2f}")
# for run this cource
# "C:\Users\rites\AppData\Local\Packages\5319275A.WhatsAppDesktop_cv1g1gvanyjgm\LocalState\sessions\E5E974812420014CBE06FD5F226D3B37E07DBD24\transfers\2026-03\app.py"
```
