months = ["January", "February", "March"]
units = []

# Input units
for m in months:
    u = int(input(f"Enter units for {m}: "))
    units.append(u)

# Slab-based bill calculation
def calculate_bill(unit):
    if unit <= 100:
        return unit * 4
    elif unit <= 200:
        return (100 * 4) + (unit - 100) * 6
    else:
        return (100 * 4) + (100 * 6) + (unit - 200) * 8

bills = [calculate_bill(u) for u in units]

# Average calculations
avg_units = sum(units) / len(units)
avg_bill = sum(bills) / len(bills)

print("\nMonthly Bills:", bills)
print("Average Units:", avg_units)
print("Average Bill: Rs", avg_bill)

# Bar Graph
plt.bar(months, bills)
plt.xlabel("Months")
plt.ylabel("Bill Amount (Rs)")
plt.title("Electricity Bill Comparison")
plt.show()

# Line Graph
plt.plot(months, units, marker='o')
plt.xlabel("Months")
plt.ylabel("Units Consumed")
plt.title("Electricity Usage Trend")
plt.show()

```python
import math
import matplotlib.pyplot as plt

# INPUT
P = float(input("Enter Loan Amount: "))
R = float(input("Enter Annual Interest Rate (%): "))
T = int(input("Enter Time (Years): "))

# SIMPLE INTEREST
SI = (P * R * T) / 100
total_si = P + SI

# COMPOUND INTEREST
CI = P * ((1 + R / 100) ** T) - P
total_ci = P + CI

# EMI CALCULATION
monthly_rate = R / (12 * 100)
months = T * 12

EMI = (P * monthly_rate * (1 + monthly_rate) ** months) / ((1 + monthly_rate) ** months - 1)
total_emi_payment = EMI * months
total_interest = total_emi_payment - P

# OUTPUT
print("\n----- LOAN DETAILS -----")
print("Loan Amount       : ‚Çπ", P)
print("Time Period       : ", T, "Years")
print("Interest Rate     : ", R, "%")

print("\nSimple Interest   : ‚Çπ", round(SI, 2))
print("Total Amount (SI) : ‚Çπ", round(total_si, 2))

print("\nCompound Interest : ‚Çπ", round(CI, 2))
print("Total Amount (CI) : ‚Çπ", round(total_ci, 2))

print("\nMonthly EMI       : ‚Çπ", round(EMI, 2))
print("Total EMI Amount  : ‚Çπ", round(total_emi_payment, 2))
print("Total Interest    : ‚Çπ", round(total_interest, 2))

# GRAPH
labels = ['Loan Amount', 'Total EMI Payment', 'Total Interest']
values = [P, total_emi_payment, total_interest]

plt.bar(labels, values)
plt.title("Bank vs Customer Loan Comparison")
plt.ylabel("Amount (‚Çπ)")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Dataset
area = np.array([500, 800, 1000, 1200, 1500])
price = np.array([25, 40, 50, 60, 75])

# Mean
x_mean = np.mean(area)
y_mean = np.mean(price)

# Slope (m)
numerator = np.sum((area - x_mean) * (price - y_mean))
denominator = np.sum((area - x_mean) ** 2)
m = numerator / denominator

# Intercept (c)
c = y_mean - m * x_mean

print("Slope (m):", m)
print("Intercept (c):", c)

# Prediction
x_new = 1100
y_pred = m * x_new + c
print("Predicted price for 1100 sq.ft:", y_pred, "Lakhs")

# Plot
plt.scatter(area, price)
plt.plot(area, m * area + c)
plt.xlabel("Area (sq.ft)")
plt.ylabel("Price (Lakhs)")
plt.title("House Price Prediction using Linear Regression")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Dataset
area = np.array([500, 800, 1000, 1200, 1500])
price = np.array([25, 40, 50, 60, 75])

# Mean of x and y
x_mean = np.mean(area)
y_mean = np.mean(price)

# Calculate slope (m)
numerator = np.sum((area - x_mean) * (price - y_mean))
denominator = np.sum((area - x_mean) ** 2)
m = numerator / denominator

# Calculate intercept (c)
c = y_mean - m * x_mean

print("Slope (m):", m)
print("Intercept (c):", c)

# Predictions for training data
y_pred = m * area + c

# Cost Function (MSE)
mse = np.mean((price - y_pred) ** 2)
print("Mean Squared Error (MSE):", mse)

# üîπ USER INPUT
user_area = float(input("Enter house area in sq.ft: "))
user_price = m * user_area + c

print("Predicted House Price:", round(user_price, 2), "Lakhs")

# Plot
plt.scatter(area, price)
plt.plot(area, y_pred)
plt.scatter(user_area, user_price)
plt.xlabel("Area (sq.ft)")
plt.ylabel("Price (Lakhs)")
plt.title("House Price Prediction using Linear Regression")
plt.show()
```

```python
import numpy as np

# Dataset
x = np.array([500, 800, 1000, 1200, 1500])
y = np.array([25, 40, 50, 60, 75])

# Initialize parameters
m = 0
c = 0
alpha = 0.0000001  # learning rate
n = len(x)

# Gradient Descent Loop
for i in range(1000):
    y_pred = m * x + c
    dm = (-2/n) * np.sum(x * (y - y_pred))
    dc = (-2/n) * np.sum(y - y_pred)

    m = m - alpha * dm
    c = c - alpha * dc

print("Final Slope (m):", round(m, 4))
print("Final Intercept (c):", round(c, 4))

```

```python
import numpy as np

# Input features: Area, Rooms
X = np.array([
    [800, 2],
    [1000, 3],
    [1200, 3],
    [1500, 4]
])

y = np.array([40, 55, 65, 80])

# Add bias column (1s)
X = np.c_[np.ones(X.shape[0]), X]

# Normal Equation
theta = np.linalg.inv(X.T @ X) @ X.T @ y

print("Intercept:", theta[0])
print("Area coefficient:", theta[1])
print("Rooms coefficient:", theta[2])

# User Input
area = float(input("Enter area: "))
rooms = int(input("Enter number of rooms: "))

price_pred = theta[0] + theta[1]*area + theta[2]*rooms
print("Predicted House Price:", round(price_pred, 2), "Lakhs")
```

```python
import numpy as np

# ==============================
# 1Ô∏è‚É£ USER‚ÄìMOVIE RATING MATRIX
# Rows = Users, Columns = Movies
# ==============================

users = np.array([
    [5, 4, 0, 0],  # User 1
    [4, 5, 0, 0],  # User 2
    [0, 0, 4, 5],  # User 3
    [0, 0, 5, 4]   # User 4
])

print("User‚ÄìMovie Rating Matrix:\n", users)

# ==============================
# 2Ô∏è‚É£ COSINE SIMILARITY FUNCTION
# ==============================

def cosine_similarity(A, B):
    return np.dot(A, B) / (np.linalg.norm(A) * np.linalg.norm(B))

# =================================================
# 3Ô∏è‚É£ USER‚ÄìUSER RECOMMENDATION SYSTEM
# =================================================

target_user_index = 0   # User 1
target_user = users[target_user_index]

print("\n--- USER‚ÄìUSER SIMILARITY ---")

user_similarities = []

for i in range(len(users)):
    if i != target_user_index:
        sim = cosine_similarity(target_user, users[i])
        user_similarities.append((i, sim))
        print(f"Similarity between User 1 and User {i+1}: {round(sim,2)}")

# Most similar user
user_similarities.sort(key=lambda x: x[1], reverse=True)
most_similar_user = user_similarities[0][0]

print("\nMost similar user to User 1: User", most_similar_user + 1)

# Recommend movies
user_user_recommendations = []

for i in range(len(target_user)):
    if target_user[i] == 0 and users[most_similar_user][i] > 0:
        user_user_recommendations.append(i+1)

print("User‚ÄìUser Recommended Movies for User 1:", user_user_recommendations)

# =================================================
# 4Ô∏è‚É£ ITEM‚ÄìITEM RECOMMENDATION SYSTEM
# =================================================

# Transpose ‚Üí Movies as rows
movies = users.T

print("\nMovie‚ÄìUser Matrix:\n", movies)

print("\n--- ITEM‚ÄìITEM SIMILARITY ---")

movie_similarity_matrix = np.zeros((movies.shape[0], movies.shape[0]))

for i in range(movies.shape[0]):
    for j in range(movies.shape[0]):
        movie_similarity_matrix[i][j] = cosine_similarity(movies[i], movies[j])

print("\nMovie Similarity Matrix:\n", np.round(movie_similarity_matrix, 2))

# Recommend movies similar to Movie 1
movie_index = 0  # Movie 1

similar_movies = list(enumerate(movie_similarity_matrix[movie_index]))
similar_movies = sorted(similar_movies, key=lambda x: x[1], reverse=True)

print("\nMovies similar to Movie 1:")
for idx, score in similar_movies[1:]:
    print(f"Movie {idx+1}  Similarity: {round(score,2)}")
```

```python
import numpy as np
# Rows = Movies, Columns = Users
movies = np.array([
    [5, 4, 0, 0],  # Movie 1
    [4, 5, 0, 0],  # Movie 2
    [0, 0, 4, 5],  # Movie 3
    [0, 0, 5, 4]   # Movie 4
])
def cosine_similarity(A, B):
    dot = np.dot(A, B)
    mag_A = np.sqrt(np.dot(A, A))
    mag_B = np.sqrt(np.dot(B, B))
    return dot / (mag_A * mag_B)

similarity_matrix = np.zeros((4,4))

for i in range(4):
    for j in range(4):
        similarity_matrix[i][j] = cosine_similarity(movies[i], movies[j])

print("Movie Similarity Matrix:\n", similarity_matrix)
movie_index = 0  # Movie 1

similar_movies = list(enumerate(similarity_matrix[movie_index]))
similar_movies = sorted(similar_movies, key=lambda x: x[1], reverse=True)

print("Recommended Movies:")
for movie, score in similar_movies[1:]:
    print("Movie", movie+1, "Similarity:", round(score, 2))
# User 1 preferences
user1 = np.array([5, 4, 0, 0])

scores = []

for i in range(len(movies)):
    score = cosine_similarity(user1, movies[i])
    scores.append(score)

print("Recommended Movie Index:", np.argmax(scores) + 1)
```
